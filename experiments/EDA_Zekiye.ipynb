## Obesity Levels Based On Eating Habits and Physical Condition

### The dataset contains 2111 rows and 17 columns, with a mix of categorical and numerical variables. Here's a breakdown of the key columns:

- Gender: Categorical (Male/Female)
- Age, Height, Weight: Numerical variables
- family_history_with_overweight, FAVC, CAEC, SMOKE, SCC, CALC, MTRANS,
-  NObeyesdad: Categorical variables representing factors such as family history of being overweight, consumption habits, physical activity, and the target obesity classification (NObeyesdad).
- FCVC, NCP, CH2O, FAF, TUE: Numerical variables representing different health metrics and habits.

```python
#First let's import the data and turn it to a dataframe

import pandas as pd

# Load the dataset
file_path = '../data/ObesityDataSet_raw_and_data_sinthetic.csv'
data = pd.read_csv(file_path)

# Display basic information about the dataset
data_info = data.info()
```

```python
data_head = data.head()
data_head
```

```python
descriptive_stats= data_describtion=data.describe()
descriptive_stats
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
# Custom color palette
palette = {'Male': 'skyblue', 'Female': 'salmon'}
# Create a count plot to show the distribution of genders across the NObeyesdad target variable
plt.figure(figsize=(10,6))
sns.countplot(x='NObeyesdad', hue='Gender', data=data, palette=palette)
plt.title('Gender Distribution across Obesity Levels (NObeyesdad)')
plt.xlabel('Obesity Level (NObeyesdad)')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Gender')
plt.tight_layout()
plt.show()
```

```python
# Boxplot shows the age distribution across different obesity levels (NObeyesdad)
palette = sns.color_palette("Set2")

# Create the boxplot with custom colors
plt.figure(figsize=(10,6))
sns.boxplot(x='NObeyesdad', y='Age', data=data, palette=palette)
plt.title('Age Distribution across Obesity Levels (NObeyesdad)')
plt.xlabel('Obesity Level (NObeyesdad)')
plt.ylabel('Age')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Boxplot shows the weight distribution across different obesity levels (NObeyesdad)
plt.figure(figsize=(10,6))
sns.boxplot(x='NObeyesdad', y='Weight', data=data, palette='Set2')
plt.title('Weight Distribution across Obesity Levels (NObeyesdad)')
plt.xlabel('Obesity Level (NObeyesdad)')
plt.ylabel('Weight')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Let's start by performing correlation analysis, pair plots, and creating distribution plots.
# Create a heatmap for the correlation matrix using only numerical columns
# Automatically select numerical columns
numerical_columns = data.select_dtypes(include=['float64', 'int64']).columns
numerical_correlation_matrix = data[numerical_columns].corr()

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(numerical_correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix of Numerical Variables')
plt.show()
```

```python
# 2. Distribution plots for continuous variables (Age, Height, Weight)
plt.figure(figsize=(14, 6))

# Age distribution
plt.subplot(1, 3, 1)
sns.histplot(data['Age'], kde=True, bins=20, color='skyblue')
plt.title('Age Distribution')

# Height distribution
plt.subplot(1, 3, 2)
sns.histplot(data['Height'], kde=True, bins=20, color='salmon')
plt.title('Height Distribution')

# Weight distribution
plt.subplot(1, 3, 3)
sns.histplot(data['Weight'], kde=True, bins=20, color='green')
plt.title('Weight Distribution')

plt.tight_layout()
plt.show()
```

```python
# 3. Pair plot to explore relationships between variables
sns.pairplot(data[['Age', 'Height', 'Weight', 'FCVC', 'CH2O', 'FAF', 'NObeyesdad']], hue='NObeyesdad', palette='Set2')
plt.show()
```

```python
# Create a scatter plot to visualize the relationship between physical activity (FAF) and weight
plt.figure(figsize=(8,6))
sns.scatterplot(x='FAF', y='Weight', data=data, hue='NObeyesdad', palette='Set2')
plt.title('Relationship between Physical Activity (FAF) and Weight')
plt.xlabel('Physical Activity Frequency (FAF)')
plt.ylabel('Weight')
plt.legend(title='Obesity Level')
plt.tight_layout()
plt.show()
```

```python
# Calculate BMI (Body Mass Index) and add it as a new column to the dataset
data['BMI'] = data['Weight'] / (data['Height'] ** 2)

# Create a scatter plot to visualize the relationship between physical activity (FAF) and BMI
plt.figure(figsize=(8,6))
sns.scatterplot(x='FAF', y='BMI', data=data, hue='NObeyesdad', palette='Set2')
plt.title('Relationship between Physical Activity (FAF) and BMI')
plt.xlabel('Physical Activity Frequency (FAF)')
plt.ylabel('BMI')
plt.legend(title='Obesity Level')
plt.tight_layout()
plt.show()
```

```python
# Create a scatter plot to visualize the relationship between water intake (CH2O) and BMI
plt.figure(figsize=(8,6))
sns.scatterplot(x='CH2O', y='BMI', data=data, hue='NObeyesdad', palette='Set2')

# Move the legend outside the plot
plt.title('Relationship between Water Intake (CH2O) and BMI')
plt.xlabel('Water Intake (CH2O)')
plt.ylabel('BMI')
plt.legend(title='Obesity Level', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Create a scatter plot to visualize the relationship between meal frequency (NCP) and BMI
plt.figure(figsize=(8,6))
sns.scatterplot(x='NCP', y='BMI', data=data, hue='NObeyesdad', palette='Set2')

# Adjusting title and labels
plt.title('Relationship between Meal Frequency (NCP) and BMI')
plt.xlabel('Meal Frequency (NCP)')
plt.ylabel('BMI')
plt.legend(title='Obesity Level', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# Create a bar plot to visualize the distribution of different target labels (NObeyesdad)
plt.figure(figsize=(10,6))
sns.countplot(x='NObeyesdad', data=data, palette='Set2')
plt.title('Distribution of Obesity Levels (NObeyesdad)')
plt.xlabel('Obesity Level')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Create a custom color palette for better distinction between transportation types
custom_palette = {
    'Walking': 'skyblue', 
    'Public_Transportation': 'salmon', 
    'Automobile': 'lightgreen', 
    'Motorbike': 'orange', 
    'Bike': 'purple'
}

# Create the updated count plot with a custom palette
plt.figure(figsize=(10,6))
sns.countplot(x='NObeyesdad', hue='MTRANS', data=data, palette=custom_palette)
plt.title('Obesity Levels vs Transportation Used')
plt.xlabel('Obesity Level (NObeyesdad)')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Transportation Used', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

```
